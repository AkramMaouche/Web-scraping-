#### Step 7 - Put everything together - Loop through results and append data inside a list

```python
model = []
mileage =[]
year = []
dealer_name =[]
price = []
city = []
for result in result_items : 
    dealer_name.append(result['dealer_name'])
    model.append(result['model'])
    year.append(result['year'])
    price.append(result['price'])
    city.append(result['city'])
    mileage.append(result['mileage'])
     
    
    
```

#### Step 8 - Pandas Dataframe - Single Page

```python
Tesla_onepage = pd.DataFrame({'Dealer_Name':dealer_name,
                             'City' : city,
                             'Model':model,
                             'Mileage' : mileage,
                             'Year':year,
                             'Price':price})
```

#### Step 9 - Scraping Multiple Pages (here: 5pages to get 100 results)

```python
model = []
mileage =[]
year = []
dealer_name =[]
price = []
city = []

for page in range(0,6) : 
    params = {
    'limit': '20',
    'make': 'Tesla',
    'page': str(page),
    'radius': '50',
    }
    response = requests.get('https://production-proxy.autolist.com/api/cwv/seo/listings', params=params)
    
    result_json = response.json()
    result_item = result_json['search_results'] 
    
    for result in result_item : 
        dealer_name.append(result['dealer_name'])
        model.append(result['model'])
        year.append(result['year'])
        price.append(result['price'])
        city.append(result['city'])
        mileage.append(result['mileage'])
    
    
```

```python
Tesla_multipage = pd.DataFrame({'Dealer_Name':dealer_name,
                             'City' : city,
                             'Model':model,
                             'Mileage' : mileage,
                             'Year':year,
                             'Price':price})
```
